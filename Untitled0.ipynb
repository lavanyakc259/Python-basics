```python

marks = int(input("Enter student marks: "))

if marks >= 50:
    print("Student is Pass")
    print("Final Marks:", marks)
else:
    grace = 50 - marks
    print("Student is Pass with Grace Marks")
    print("Grace Marks Added:", grace)
    print("Final Marks:", 50)
```
